```python
"""--------------------------------------------------------------------------
    1. Momento del d√≠a en el que m√°s compran determinado tipo de clientes
   --------------------------------------------------------------------------
"""
df['Date/Time'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
del(df['Time'])
del(df['Date'])
df.head()
```

```python
df['Hour'] = df['Date/Time'].dt.hour
df.head()
```

```python
df[df['Customer type'] == 'Member']['Hour'].hist(alpha=.5)
df[df['Customer type'] == 'Normal']['Hour'].hist(alpha=.5)
plt.legend(['Member', 'Normal'])
plt.show()
```

```python
"""--------------------------------------------------------------------------
    2. ¬øQu√© compran m√°s hombres y mujeres en determinadas fechas del a√±o? 
   --------------------------------------------------------------------------
"""
gender_vs_productLine_avg_sells = df.pivot_table('Total', index='Gender', columns='Product line')
gender_vs_productLine_avg_sells
```

```python
# sns.catplot(x="Gender", y="Total", hue="Product line", kind="bar", data=df) -> No es muy clara la gr√°fica...
labels = gender_vs_productLine_avg_sells.columns
men_means = gender_vs_productLine_avg_sells.loc['Male']
women_means = gender_vs_productLine_avg_sells.loc['Female']

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()
fig.set_size_inches(18.5, 10.5)
# plt.xticks(rotation=45)
plt.show()
```
