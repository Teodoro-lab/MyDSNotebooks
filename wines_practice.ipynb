```python
def ecdf(data):
    x_values = np.sort(data)
    n = len(data)
    y_values = np.arange(1, n + 1) / n
    return x_values, y_values
```

```python
np.arange(1, 10)/9
```

```python
prices, percentage = ecdf(df['price'])
```

```python
sns.set()
plt.plot(prices, percentage, marker='.', linestyle='none')
plt.xlabel('prices')
plt.ylabel('ECDF')
```
