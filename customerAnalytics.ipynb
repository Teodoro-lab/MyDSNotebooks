```python
train_data = pd.read_csv('/kaggle/input/customer-analytics/Train.csv')
train_data.head()
```

```python
customer_rating_vs_reachedOnTime_sorted = train_data.groupby('Reached.on.Time_Y.N')['Customer_rating'].value_counts().sort_index(ascending=False)
```

```python
from matplotlib import pyplot as plt
plt.plot(customer_rating_vs_reachedOnTime_sorted.loc[1].index, customer_rating_vs_reachedOnTime_sorted.loc[1].values, kind='bar')
plt.plot(customer_rating_vs_reachedOnTime_sorted.loc[0].index, customer_rating_vs_reachedOnTime_sorted.loc[0].values, kind='bar')
plt.show()
```
